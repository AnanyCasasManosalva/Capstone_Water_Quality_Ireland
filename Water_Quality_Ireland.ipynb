# Analize some pameters for each selected river

### 62 rivers were selected and information was found in 46 of them, so all of them will be joined in a DataFrame to perform the correponding EDA (a file had to be downloaded for each river on the EPA website)

```python
import os

folder_path = "Dataset_selected/Rivers_Tipperary"
files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

data_frames = []
for file in files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    data_frames.append(df)

# Unir todos los DataFrames en uno solo
combined_df = pd.concat(data_frames, ignore_index=True)
combined_df
```

### It can be seen see that some values of MonitoringStationLocalAuthority include several names, let's see which ones they contain since we only want to analyze 'Tipperary County Council'

```python
# Display the unique names in the column 'MonitoringStationLocalAuthority'
unique_numbers = combined_df['MonitoringStationLocalAuthority'].unique()

# Display the unique names
print(unique_numbers)
```

```python
# Filter the DataFrame with the name 'Tipperary County Council' in the column 'MonitoringStationLocalAuthority'
combined_df_tipperary = combined_df[combined_df['MonitoringStationLocalAuthority'] == 'Tipperary County Council']
combined_df_tipperary
```

```python
combined_df_tipperary.info()
```

```python
unique_rivers = combined_df_tipperary['WaterbodyName'].unique()
print(unique_rivers)
```

```python
print(f"\nUnique rivers in the area of Tipperary County Council with rivers found:, {len(unique_rivers)}")
```

```python
# Remove unnecessary columns
columns_to_drop = [
    'WaterbodyCode', 'MonitoringStationCode', 'MonitoringStationType', 'SampleCode', 'SampleMethod', 'ParameterUnitName', 'Result',
    'TextResult', 'ResultString', 'LimitOfDetection', 'ReportTextResult', 'ReportResultString', 'ReportLimit'
]

combined_df_tipperary.drop(columns=columns_to_drop, inplace=True)
combined_df_tipperary
```

```python
# Show information about null values in DataFrame columns
print(combined_df_tipperary.isnull().sum())
```

```python
# Calculating the percentage of null values
total_rows = 110481
null_values = 4140
percentage_nulls = (null_values / total_rows) * 100
print(f"Percentage of null values: {percentage_nulls:.2f}%")
```

## CASE ONE: Remove rows with null values

### 4140 null values represents 3.75% of 110481 which is withing the accepted threshold of 5%, so rows with null values will be eliminated

```python
# Remove the rows with null values in 'Result' column
combined_df_tipperary.dropna(subset=['ReportResult'], inplace=True)
combined_df_tipperary
```

```python
# Save the DataFrame as a CSV file
combined_df_tipperary.to_csv('combined_df_tipperary.csv', index=False)
```
